# Exploratory Data Analysis

```python
gender = sns.countplot(x="Gender", data=df)
for label in gender.containers:
    gender.bar_label(label)
```

```python
filtered_gender = (
    df.groupby("Gender", as_index=False)["Amount"]
    .sum()
    .sort_values(by="Amount", ascending=False)
)
```

```python
sns.barplot(x="Gender", y="Amount", data=filtered_gender)
```

### From the graphs we can see that most of the orders are placed by women 

```python
ax = sns.countplot(x='Age Group', hue='Gender', data=df)
for label in ax.containers:
    ax.bar_label(label)
```

```python
filtered_age_group = df.groupby('Age Group', as_index=False)['Amount'].sum().sort_values(ascending=False, by='Amount')
sns.barplot(x='Age Group', y='Amount', data=filtered_age_group)
```

```python
filtered_state = df.groupby('State', as_index=False)['Orders'].sum().sort_values(ascending=False, by='Orders')
filtered_state
sns.set(rc={'figure.figsize':(25,8)})
sns.barplot(x='State', y='Orders', data=filtered_state)
```

```python
filtered_state_amount = df.groupby('State', as_index=False)['Amount'].sum().sort_values(ascending=False, by='Amount')
sns.set(rc={'figure.figsize':(25,8)})
sns.barplot(x='State', y='Amount', data=filtered_state_amount)
```

```python
marital = sns.countplot(x='Shaadi', data=df)
sns.set(rc={'figure.figsize': (7,5)})
for label in marital.containers:
    marital.bar_label(label)
```

```python
filtered_state_married = df.groupby(['Shaadi', 'Gender'], as_index=False)['Amount'].sum().sort_values(ascending=False, by='Amount')
sns.set(rc={'figure.figsize': (9,5)})
sns.barplot(x='Shaadi', y='Amount', hue='Gender' ,data=filtered_state_married)
```

```python
heda = sns.countplot(x='Occupation', data=df)
sns.set(rc={'figure.figsize':(28,6)})
for label in heda.containers:
    heda.bar_label(label)
```

```python
amount = df.groupby('Occupation', as_index=False)['Amount'].sum().sort_values(ascending=False, by='Amount')
sns.barplot(x='Occupation', y='Amount', data=amount)
```

```python
amount_up = df.groupby(['Occupation', 'Gender'], as_index=False)['Amount'].sum().sort_values(ascending=False, by='Amount')
sns.barplot(x='Occupation', y='Amount', hue='Gender',data=amount_up)
```
